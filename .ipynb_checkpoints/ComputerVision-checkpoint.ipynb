Any digital image can be considered a function. So an image essentially has all the properties of a function and operations related to functions can be performed on images.

The following example (using openCV and python) explains how:    

<i><b>Note:</b> The functions like cvtColor, COLOR_BGR2RGB which are used here, will be explained in detail in a section which comes later</i>

```python
import cv2 as cv
import matplotlib.pyplot as plt
##READ THE IMAGE
##The function imread() is used to read the image 
##The argument 0 is used to read the image in greyscale mode
linusImageColor = cv.imread("linus-torvalds.jpg",0)
print ("The data type is: ",type(linusImageColor))
plt.imshow(linusImageColor,cmap="gray")
```

As we can see, the datatype is a numpy.ndarray which is an [array](htps://en.wikipedia.org/wiki/Array_data_structure).    
We can also plot this as a function,which is done in the example below which uses MATLAB:
